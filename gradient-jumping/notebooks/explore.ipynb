```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)
z = x**2 + y**2  # Equation of upwards parabola

# Choose points on the surface
point_index1 = (25, 25)  # Indices of the first point in the grid
point_index2 = (25, 75)  # Indices of the second point in the grid
point_index3 = (60, 50)  # Indices of the second point in the grid
point1 = (x[point_index1], y[point_index1], z[point_index1])
point2 = (x[point_index2], y[point_index2], z[point_index2])
point3 = (x[point_index3], y[point_index3], z[point_index3])

# Calculate the equation of the tangent planes at the chosen points
# The equation of the tangent plane at point (x0, y0, z0) is given by:
# z = f(x0, y0) + f_x(x0, y0) * (x - x0) + f_y(x0, y0) * (y - y0)
# Where f_x and f_y are the partial derivatives of f(x, y) with respect to x and y, respectively.
f_x1 = 2 * point1[0]
f_y1 = 2 * point1[1]
z_tangent_plane1 = point1[2] + f_x1 * (x - point1[0]) + f_y1 * (y - point1[1])

f_x2 = 2 * point2[0]
f_y2 = 2 * point2[1]
z_tangent_plane2 = point2[2] + f_x2 * (x - point2[0]) + f_y2 * (y - point2[1])

f_x3 = 2 * point3[0]
f_y3 = 2 * point3[1]
z_tangent_plane3 = point3[2] + f_x3 * (x - point3[0]) + f_y3 * (y - point3[1])

# Find the line of intersection between the two planes
# Solve the system of equations formed by the equations of the two planes
# z_tangent_plane1 = z_tangent_plane2
# This will give the equation of the line of intersection
# z_tangent_plane1 - z_tangent_plane2 = 0
line_intersection_12 = np.where(np.isclose(z_tangent_plane1 - z_tangent_plane2, 0))
line_intersection_23 = np.where(np.isclose(z_tangent_plane2 - z_tangent_plane3, 0))
line_intersection_13 = np.where(np.isclose(z_tangent_plane1 - z_tangent_plane3, 0))

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot surface
ax.plot_surface(x, y, z, cmap='viridis')

# Plot tangent planes
ax.plot_surface(x, y, z_tangent_plane1, alpha=0.5, color='orange')
ax.plot_surface(x, y, z_tangent_plane2, alpha=0.5, color='green')
ax.plot_surface(x, y, z_tangent_plane3, alpha=0.5, color='green')

# Plot line of intersection
ax.plot(x[line_intersection_12], y[line_intersection_12], z_tangent_plane1[line_intersection_12], color='red')
ax.plot(x[line_intersection_23], y[line_intersection_23], z_tangent_plane2[line_intersection_23], color='red')
ax.plot(x[line_intersection_13], y[line_intersection_13], z_tangent_plane1[line_intersection_13], color='red')

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Adjust viewing angle
ax.view_init(elev=20, azim=135)

plt.show()
```
